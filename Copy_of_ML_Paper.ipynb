```python
# Get the count of missing values
missing_values = df_cleaned.isnull().sum()
total_rows = len(df_cleaned)  # Get the total number of rows

# Create a formatted output
print("\nMissing values count with percentage:")
print(f"{'Index':<6} {'Column Name':<50} {'Missing Values':<20} {'Percentage Missing':<20}")

# Display the count and percentage of missing values for each column
for index, column in enumerate(df_cleaned.columns):
    count_missing = missing_values[column]
    percentage_missing = (count_missing / total_rows) * 100  # Calculate percentage
    print(f"{index + 1:<6} {column:<50} {count_missing:<20} {percentage_missing:<20.2f}")
```

###display columns with more than 90% missing value

```python
# Get the total number of rows
total_rows = len(df_cleaned)

# Calculate the percentage of missing values for each column
missing_percentage = (df_cleaned.isnull().sum() / total_rows) * 100

# Filter columns with more than 90% missing values
columns_above_90_missing = missing_percentage[missing_percentage > 90]

# Display columns with more than 90% missing values (starting index from 1)
print("\nColumns with more than 90% missing values:")
print(f"{'Index':<6} {'Column Name':<50} {'Percentage Missing':<20}")
for index, (column, percentage) in enumerate(columns_above_90_missing.items(), start=1):
    print(f"{index:<6} {column:<50} {percentage:<20.2f}")
```

##**Identify the column contents**

```python
import pandas as pd

def analyze_column(df_cleaned, column_name):
    # Get the unique values in the column
    unique_values = df_cleaned[column_name].unique()
    print("Unique values:")
    print(unique_values)

    # Count the occurrences of each unique value
    value_counts = df_cleaned[column_name].value_counts()
    print("\nValue counts:")
    print(value_counts)

    # Get the number of unique values
    num_unique = df_cleaned[column_name].nunique()
    print("\nNumber of unique values:")
    print(num_unique)

    # Get the total number of values in the column (including duplicates)
    total_count = len(df_cleaned[column_name])
    print("\nTotal number of values:")
    print(total_count)

    # Count the number of NaN (Not a Number) values
    nan_count = df_cleaned[column_name].isnull().sum()
    print("\nNumber of NaN values:")
    print(nan_count)

# Example usage
# df = pd.DataFrame(...)  # استبدل هذه السطر بإنشاء DataFrame الخاص بك
# analyze_column(df, 'vital_status')
```
